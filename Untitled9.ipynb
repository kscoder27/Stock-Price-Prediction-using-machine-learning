```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, GRU, Dropout
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')

# Load and preprocess the dataset
df = pd.read_csv('S&P500HistoricalData.csv')
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')
df['Price'] = df['Price'].str.replace(',', '').astype(float)
data = df['Price'].values.reshape(-1, 1)

# Prepare the data
def prepare_data(data, sequence_length=60):
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(data)
    X, y = [], []
    for i in range(len(scaled_data) - sequence_length):
        X.append(scaled_data[i:i + sequence_length])
        y.append(scaled_data[i + sequence_length])
    X, y = np.array(X), np.array(y)
    train_size = int(len(X) * 0.8)
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]
    return X_train, X_test, y_train, y_test, scaler

sequence_length = 60
X_train, X_test, y_train, y_test, scaler = prepare_data(data, sequence_length)

# Split training data into training and validation sets
val_size = int(len(X_train) * 0.1)
X_val, y_val = X_train[-val_size:], y_train[-val_size:]
X_train, y_train = X_train[:-val_size], y_train[:-val_size]

# Build the LSTM+GRU model
def build_lstm_gru_model(input_shape):
    model = Sequential()
    model.add(LSTM(50, return_sequences=True, input_shape=input_shape))
    model.add(Dropout(0.2))
    model.add(GRU(50, return_sequences=False))
    model.add(Dropout(0.2))
    model.add(Dense(25))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')
    return model

model = build_lstm_gru_model((sequence_length, 1))
history = model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=16,
    validation_data=(X_val, y_val),
    verbose=1
)

# Make predictions for validation and test sets
val_predictions = model.predict(X_val)
test_predictions = model.predict(X_test)

# Inverse transform the predictions and true values
y_val_inv = scaler.inverse_transform(y_val)
val_predictions_inv = scaler.inverse_transform(val_predictions)
y_test_inv = scaler.inverse_transform(y_test)
test_predictions_inv = scaler.inverse_transform(test_predictions)

# Calculate metrics for validation and test sets
def calculate_metrics(true, pred):
    rmse = np.sqrt(mean_squared_error(true, pred))
    mae = mean_absolute_error(true, pred)
    r2 = r2_score(true, pred)
    mape = np.mean(np.abs((true - pred) / true)) * 100
    return rmse, mae, r2, mape

val_metrics = calculate_metrics(y_val_inv, val_predictions_inv)
test_metrics = calculate_metrics(y_test_inv, test_predictions_inv)

# Print metrics
print("\nValidation Set Metrics:")
print(f"RMSE: ${val_metrics[0]:.2f}")
print(f"MAE: ${val_metrics[1]:.2f}")
print(f"MAPE: {val_metrics[3]:.2f}%")
print(f"R-squared: {val_metrics[2]:.4f}")

print("\nTest Set Metrics:")
print(f"RMSE: ${test_metrics[0]:.2f}")
print(f"MAE: ${test_metrics[1]:.2f}")
print(f"MAPE: {test_metrics[3]:.2f}%")
print(f"R-squared: {test_metrics[2]:.4f}")

# Plot validation and test predictions vs actual
# Import necessary libraries
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, GRU, Dropout, Bidirectional
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# Load and preprocess the dataset
df = pd.read_csv('S&P500HistoricalData.csv')
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')
df['Price'] = df['Price'].str.replace(',', '').astype(float)
data = df['Price'].values.reshape(-1, 1)

# Prepare the data
def prepare_data(data, sequence_length=60):
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(data)
    X, y = [], []
    for i in range(len(scaled_data) - sequence_length):
        X.append(scaled_data[i:i + sequence_length])
        y.append(scaled_data[i + sequence_length])
    X, y = np.array(X), np.array(y)

    train_size = int(len(X) * 0.7)  # 70% for training
    val_size = int(len(X) * 0.15)  # 15% for validation
    X_train, X_val, X_test = X[:train_size], X[train_size:train_size + val_size], X[train_size + val_size:]
    y_train, y_val, y_test = y[:train_size], y[train_size:train_size + val_size], y[train_size + val_size:]
    return X_train, X_val, X_test, y_train, y_val, y_test, scaler

sequence_length = 60
X_train, X_val, X_test, y_train, y_val, y_test, scaler = prepare_data(data, sequence_length)

# Build the BiLSTM+GRU model
def build_bilstm_gru_model(input_shape):
    model = Sequential()
    model.add(Bidirectional(LSTM(50, return_sequences=True), input_shape=input_shape))
    model.add(Dropout(0.2))
    model.add(GRU(50, return_sequences=False))
    model.add(Dropout(0.2))
    model.add(Dense(25))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')
    return model

model = build_bilstm_gru_model((sequence_length, 1))
history = model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=16,
    validation_data=(X_val, y_val),
    verbose=1
)

# Make predictions
predictions_val = model.predict(X_val)
predictions_test = model.predict(X_test)
y_val_inv = scaler.inverse_transform(y_val)
y_test_inv = scaler.inverse_transform(y_test)
predictions_val_inv = scaler.inverse_transform(predictions_val)
predictions_test_inv = scaler.inverse_transform(predictions_test)

# Calculate metrics
def calculate_metrics(y_true, y_pred):
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    return rmse, mae, r2, mape

val_metrics = calculate_metrics(y_val_inv, predictions_val_inv)
test_metrics = calculate_metrics(y_test_inv, predictions_test_inv)

# Model Loss Convergence curve (only one plot)
train_loss = history.history['loss']
val_loss = history.history['val_loss']
smoothed_val_loss = smooth_curve(val_loss)

plt.figure(figsize=(10, 6))
plt.plot(train_loss, label='Training Loss', linestyle='-', marker='o', color='blue')
plt.plot(smoothed_val_loss, label='Validation Loss (Smoothed)', linestyle='--', marker='x', color='orange')
plt.title('Model Loss Convergence(BiLSTM+GRU)', fontsize=16, fontweight='bold')
plt.xlabel('Epochs', fontsize=12)
plt.ylabel('Loss', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.axhline(y=min(history.history['val_loss']), color='green', linestyle='--', label='Min Validation Loss')
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Print metrics
print("\nValidation Set Metrics:")
print(f"RMSE: ${val_metrics[0]:.2f}")
print(f"MAE: ${val_metrics[1]:.2f}")
print(f"MAPE: {val_metrics[3]:.2f}%")
print(f"R-squared: {val_metrics[2]:.4f}")

print("\nTest Set Metrics:")
print(f"RMSE: ${test_metrics[0]:.2f}")
print(f"MAE: ${test_metrics[1]:.2f}")
print(f"MAPE: {test_metrics[3]:.2f}%")
print(f"R-squared: {test_metrics[2]:.4f}")

# Plot for Validation and Test Predictions vs Actual
plt.figure(figsize=(15, 7))
plt.plot(df['Date'].iloc[-len(y_val_inv) - len(y_test_inv):-len(y_test_inv)], y_val_inv, label='Validation Actual', marker='o')
plt.plot(df['Date'].iloc[-len(y_val_inv) - len(y_test_inv):-len(y_test_inv)], predictions_val_inv, label='Validation Predicted', marker='o')
plt.plot(df['Date'].iloc[-len(y_test_inv):], y_test_inv, label='Test Actual', marker='o')
plt.plot(df['Date'].iloc[-len(y_test_inv):], predictions_test_inv, label='Test Predicted', marker='o')
plt.title('Validation and Test Predictions vs Actual(BiLSTM+GRU)', fontsize=16, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Regression Graph for Validation and Test Sets
plt.figure(figsize=(10, 8))
plt.scatter(y_val_inv, predictions_val_inv, alpha=0.7, c='blue', edgecolor='black', label='Validation Predictions')
plt.scatter(y_test_inv, predictions_test_inv, alpha=0.7, c='purple', edgecolor='black', label='Test Predictions')
plt.plot(
    [min(y_val_inv.min(), y_test_inv.min()), max(y_val_inv.max(), y_test_inv.max())],
    [min(y_val_inv.min(), y_test_inv.min()), max(y_val_inv.max(), y_test_inv.max())],
    'r--', label='Ideal Fit'
)
plt.title('Regression Graph for Validation and Test Sets(BiLSTM+GRU)', fontsize=16, fontweight='bold')
plt.xlabel('Actual Prices', fontsize=12)
plt.ylabel('Predicted Prices', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, GRU, Dropout, Bidirectional
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

# Load and preprocess the dataset
df = pd.read_csv('S&P500HistoricalData.csv')
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')
df['Price'] = df['Price'].str.replace(',', '').astype(float)
data = df['Price'].values.reshape(-1, 1)

# Prepare the data
def prepare_data(data, sequence_length=60):
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(data)
    X, y = [], []
    for i in range(len(scaled_data) - sequence_length):
        X.append(scaled_data[i:i + sequence_length])
        y.append(scaled_data[i + sequence_length])
    X, y = np.array(X), np.array(y)

    train_size = int(len(X) * 0.7)  # 70% for training
    val_size = int(len(X) * 0.15)   # 15% for validation
    X_train, X_val, X_test = X[:train_size], X[train_size:train_size + val_size], X[train_size + val_size:]
    y_train, y_val, y_test = y[:train_size], y[train_size:train_size + val_size], y[train_size + val_size:]
    return X_train, X_val, X_test, y_train, y_val, y_test, scaler

def smooth_curve(points, factor=0.8):
    smoothed_points = []
    for point in points:
        if smoothed_points:
            previous = smoothed_points[-1]
            smoothed_points.append(previous * factor + point * (1 - factor))
        else:
            smoothed_points.append(point)
    return smoothed_points

# Set sequence length and prepare data
sequence_length = 60
X_train, X_val, X_test, y_train, y_val, y_test, scaler = prepare_data(data, sequence_length)

# Build the BiLSTM+GRU model
def build_bilstm_gru_model(input_shape):
    model = Sequential()
    model.add(Bidirectional(LSTM(50, return_sequences=True), input_shape=input_shape))
    model.add(Dropout(0.2))
    model.add(GRU(50, return_sequences=False))
    model.add(Dropout(0.2))
    model.add(Dense(25))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')
    return model

# Create and train the model
model = build_bilstm_gru_model((sequence_length, 1))
history = model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=16,
    validation_data=(X_val, y_val),
    verbose=1
)

# Make predictions
predictions_val = model.predict(X_val)
predictions_test = model.predict(X_test)

# Transform predictions back to original scale
y_val_inv = scaler.inverse_transform(y_val)
y_test_inv = scaler.inverse_transform(y_test)
predictions_val_inv = scaler.inverse_transform(predictions_val)
predictions_test_inv = scaler.inverse_transform(predictions_test)

# Calculate metrics function
def calculate_metrics(y_true, y_pred, scaled=True):
    """
    Calculate metrics either in scaled (0-1) or original scale
    """
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100

    if scaled:
        return rmse, mae, r2, mape
    else:
        return rmse, mae, r2, mape

# Calculate metrics on both scaled and original data
scaled_val_metrics = calculate_metrics(y_val, predictions_val)
scaled_test_metrics = calculate_metrics(y_test, predictions_test)
original_val_metrics = calculate_metrics(y_val_inv, predictions_val_inv, scaled=False)
original_test_metrics = calculate_metrics(y_test_inv, predictions_test_inv, scaled=False)

# Print metrics
print("\nValidation Set Metrics:")
print("Scaled (0-1) Metrics:")
print(f"RMSE: {scaled_val_metrics[0]:.4f}")
print(f"MAE: {scaled_val_metrics[1]:.4f}")
print(f"MAPE: {scaled_val_metrics[3]:.2f}%")
print(f"R-squared: {scaled_val_metrics[2]:.4f}")

print("\nOriginal Scale Metrics:")
print(f"RMSE: ${original_val_metrics[0]:.2f}")
print(f"MAE: ${original_val_metrics[1]:.2f}")
print(f"MAPE: {original_val_metrics[3]:.2f}%")
print(f"R-squared: {original_val_metrics[2]:.4f}")

print("\nTest Set Metrics:")
print("Scaled (0-1) Metrics:")
print(f"RMSE: {scaled_test_metrics[0]:.4f}")
print(f"MAE: {scaled_test_metrics[1]:.4f}")
print(f"MAPE: {scaled_test_metrics[3]:.2f}%")
print(f"R-squared: {scaled_test_metrics[2]:.4f}")

print("\nOriginal Scale Metrics:")
print(f"RMSE: ${original_test_metrics[0]:.2f}")
print(f"MAE: ${original_test_metrics[1]:.2f}")
print(f"MAPE: {original_test_metrics[3]:.2f}%")
print(f"R-squared: {original_test_metrics[2]:.4f}")

# Plotting
# 1. Model Loss Convergence curve
train_loss = history.history['loss']
val_loss = history.history['val_loss']
smoothed_val_loss = smooth_curve(val_loss)

plt.figure(figsize=(10, 6))
plt.plot(train_loss, label='Training Loss', linestyle='-', marker='o', color='blue')
plt.plot(smoothed_val_loss, label='Validation Loss (Smoothed)', linestyle='--', marker='x', color='orange')
plt.title('Model Loss Convergence (BiLSTM+GRU)', fontsize=16, fontweight='bold')
plt.xlabel('Epochs', fontsize=12)
plt.ylabel('Loss', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.axhline(y=min(history.history['val_loss']), color='green', linestyle='--', label='Min Validation Loss')
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# 2. Validation and Test Predictions vs Actual
plt.figure(figsize=(15, 7))
plt.plot(df['Date'].iloc[-len(y_val_inv) - len(y_test_inv):-len(y_test_inv)],
         y_val_inv, label='Validation Actual', marker='o')
plt.plot(df['Date'].iloc[-len(y_val_inv) - len(y_test_inv):-len(y_test_inv)],
         predictions_val_inv, label='Validation Predicted', marker='o')
plt.plot(df['Date'].iloc[-len(y_test_inv):],
         y_test_inv, label='Test Actual', marker='o')
plt.plot(df['Date'].iloc[-len(y_test_inv):],
         predictions_test_inv, label='Test Predicted', marker='o')
plt.title('Validation and Test Predictions vs Actual (BiLSTM+GRU)', fontsize=16, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# 3. Regression Graph
plt.figure(figsize=(10, 8))
plt.scatter(y_val_inv, predictions_val_inv, alpha=0.7, c='blue',
           edgecolor='black', label='Validation Predictions')
plt.scatter(y_test_inv, predictions_test_inv, alpha=0.7, c='purple',
           edgecolor='black', label='Test Predictions')
plt.plot(
    [min(y_val_inv.min(), y_test_inv.min()), max(y_val_inv.max(), y_test_inv.max())],
    [min(y_val_inv.min(), y_test_inv.min()), max(y_val_inv.max(), y_test_inv.max())],
    'r--', label='Ideal Fit'
)
plt.title('Regression Graph for Validation and Test Sets (BiLSTM+GRU)', fontsize=16, fontweight='bold')
plt.xlabel('Actual Prices', fontsize=12)
plt.ylabel('Predicted Prices', fontsize=12)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, GRU, Dropout
import matplotlib.pyplot as plt
import warnings

warnings.filterwarnings('ignore')

# Load and preprocess the dataset
df = pd.read_csv('S&P500HistoricalData.csv')
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values('Date')
df['Price'] = df['Price'].str.replace(',', '').astype(float)
data = df['Price'].values.reshape(-1, 1)

# Prepare the data
def prepare_data(data, sequence_length=60):
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(data)
    X, y = [], []
    for i in range(len(scaled_data) - sequence_length):
        X.append(scaled_data[i:i + sequence_length])
        y.append(scaled_data[i + sequence_length])
    X, y = np.array(X), np.array(y)
    train_size = int(len(X) * 0.8)
    X_train, X_test = X[:train_size], X[train_size:]
    y_train, y_test = y[:train_size], y[train_size:]
    return X_train, X_test, y_train, y_test, scaler

sequence_length = 60
X_train, X_test, y_train, y_test, scaler = prepare_data(data, sequence_length)

# Split training data into training and validation sets
val_size = int(len(X_train) * 0.1)
X_val, y_val = X_train[-val_size:], y_train[-val_size:]
X_train, y_train = X_train[:-val_size], y_train[:-val_size]

# Build the LSTM+GRU model
def build_lstm_gru_model(input_shape):
    model = Sequential()
    model.add(LSTM(50, return_sequences=True, input_shape=input_shape))
    model.add(Dropout(0.2))
    model.add(GRU(50, return_sequences=False))
    model.add(Dropout(0.2))
    model.add(Dense(25))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mse')
    return model

model = build_lstm_gru_model((sequence_length, 1))
history = model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=16,
    validation_data=(X_val, y_val),
    verbose=1
)

# Make predictions for validation and test sets
val_predictions = model.predict(X_val)
test_predictions = model.predict(X_test)

# Inverse transform the predictions and true values
y_val_inv = scaler.inverse_transform(y_val)
val_predictions_inv = scaler.inverse_transform(val_predictions)
y_test_inv = scaler.inverse_transform(y_test)
test_predictions_inv = scaler.inverse_transform(test_predictions)

# Calculate metrics for validation and test sets
def calculate_metrics(true, pred):
    rmse = np.sqrt(mean_squared_error(true, pred))
    mae = mean_absolute_error(true, pred)
    r2 = r2_score(true, pred)
    mape = np.mean(np.abs((true - pred) / true)) * 100
    return rmse, mae, r2, mape

val_metrics = calculate_metrics(y_val_inv, val_predictions_inv)
test_metrics = calculate_metrics(y_test_inv, test_predictions_inv)

# Print metrics
print("\nValidation Set Metrics:")
print(f"RMSE: ${val_metrics[0]:.2f}")
print(f"MAE: ${val_metrics[1]:.2f}")
print(f"MAPE: {val_metrics[3]:.2f}%")
print(f"R-squared: {val_metrics[2]:.4f}")

print("\nTest Set Metrics:")
print(f"RMSE: ${test_metrics[0]:.2f}")
print(f"MAE: ${test_metrics[1]:.2f}")
print(f"MAPE: {test_metrics[3]:.2f}%")
print(f"R-squared: {test_metrics[2]:.4f}")

# Plot validation and test predictions vs actual
plt.figure(figsize=(15, 7))
plt.plot(df['Date'].iloc[-len(y_val_inv) - len(y_test_inv):-len(y_test_inv)], y_val_inv, label='Validation Actual', marker='o')
plt.plot(df['Date'].iloc[-len(y_val_inv) - len(y_test_inv):-len(y_test_inv)], val_predictions_inv, label='Validation Predicted', marker='o')
plt.plot(df['Date'].iloc[-len(y_test_inv):], y_test_inv, label='Test Actual', marker='o')
plt.plot(df['Date'].iloc[-len(y_test_inv):], test_predictions_inv, label='Test Predicted', marker='o')
plt.title('Validation and Test Predictions vs Actual(LSTM+GRU)', fontsize=16, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Regression graph for validation and test sets
plt.figure(figsize=(10, 8))
plt.scatter(y_val_inv, val_predictions_inv, alpha=0.7, c='blue', edgecolor='black', label='Validation Predictions')
plt.scatter(y_test_inv, test_predictions_inv, alpha=0.7, c='purple', edgecolor='black', label='Test Predictions')
plt.plot(
    [min(y_val_inv.min(), y_test_inv.min()), max(y_val_inv.max(), y_test_inv.max())],
    [min(y_val_inv.min(), y_test_inv.min()), max(y_val_inv.max(), y_test_inv.max())],
    'r--',
    lw=2,
    label='Ideal Fit'
)
plt.title('Regression Graph: Validation and Test Sets(LSTM+GRU)', fontsize=16, fontweight='bold')
plt.xlabel('Actual Prices', fontsize=12)
plt.ylabel('Predicted Prices', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Improved convergence curve
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Training Loss', linestyle='-', marker='o', color='blue')
plt.plot(history.history['val_loss'], label='Validation Loss', linestyle='--', marker='x', color='orange')
plt.title('Model Loss Convergence(LSTM+GRU)', fontsize=16, fontweight='bold')
plt.xlabel('Epochs', fontsize=12)
plt.ylabel('Loss', fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.axhline(y=min(history.history['val_loss']), color='green', linestyle='--', label='Min Validation Loss')
plt.legend(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
